```python
# Import the dependency.
import pandas as pd
```

```python
 # Set the column width
pd.options.display.max_colwidth = 200
#  Read in the dataset. 
trips_df = pd.read_csv("Resources/may23_citibike.csv")
trips_df.head()
```

```python
trips_df.info()
```

```python
# Convert datatype in started at to datetime
trips_df['started_at'] = pd.to_datetime(trips_df['started_at'])
# Convert datatype in ended at and ended at to datetime
trips_df['ended_at'] = pd.to_datetime(trips_df['ended_at'])

trips_df.info()
```

```python
 # Rename the columns. 
trips_df.rename(columns={'started_at': "starting date_time",
                            'ended_at': "ending date_time",
                            'rideable_type': "bike_type",
                            'member_casual': 'membership_or_casual'}, inplace=True)
trips_df.head()
```

```python
trips_df.count()
```

```python
# Drop all rows with missing information
trips_df = trips_df.dropna(how='any')
```

```python
trips_df.count()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
trips_df.to_csv("Resources/cleanedData.csv", index=False, header=True)
```
